# Define FFT and IFFT Functions
Define the FFT and IFFT functions using NumPy.

```python
import cmath

# Define the FFT function manually
def fft(signal):
    N = len(signal)
    if N <= 1:
        return signal
    even = fft(signal[0::2])
    odd = fft(signal[1::2])
    T = [cmath.exp(-2j * cmath.pi * k / N) * odd[k] for k in range(N // 2)]
    return [even[k] + T[k] for k in range(N // 2)] + [even[k] - T[k] for k in range(N // 2)]

# Define the IFFT function manually
def ifft(signal):
    N = len(signal)
    if N <= 1:
        return signal
    even = ifft(signal[0::2])
    odd = ifft(signal[1::2])
    T = [cmath.exp(2j * cmath.pi * k / N) * odd[k] for k in range(N // 2)]
    return [(even[k] + T[k]) / 2 for k in range(N // 2)] + [(even[k] - T[k]) / 2 for k in range(N // 2)]
```

# Mathematical Formulation of FFT and IFFT

In this section, we will provide the mathematical formulas for the Fast Fourier Transform (FFT) and its inverse (IFFT) using LaTeX.

## Fast Fourier Transform (FFT)

The FFT is an algorithm to compute the Discrete Fourier Transform (DFT) efficiently. The DFT of a sequence \( x[n] \) of length \( N \) is given by:

$$
X[k] = \sum_{n=0}^{N-1} x[n] \cdot e^{-j \frac{2\pi}{N} kn}, \quad k = 0, 1, \ldots, N-1
$$

## Inverse Fast Fourier Transform (IFFT)

The IFFT is used to compute the inverse of the DFT. The inverse DFT of a sequence \( X[k] \) of length \( N \) is given by:

$$
x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] \cdot e^{j \frac{2\pi}{N} kn}, \quad n = 0, 1, \ldots, N-1
$$

# Generate Sample Signal
Generate a sample signal (e.g., a sine wave) for demonstration purposes.

```python
# Generate Sample Signal

import numpy as np
import matplotlib.pyplot as plt

# Generate a sample signal (sine wave)
sampling_rate = 1000  # Sampling rate in Hz
t = np.linspace(0, 1, sampling_rate, endpoint=False)  # Time vector
frequency = 5  # Frequency of the sine wave in Hz
amplitude = 1  # Amplitude of the sine wave

# Create the sine wave signal
signal = amplitude * np.sin(2 * np.pi * frequency * t)

# Plot the sample signal
plt.figure(figsize=(10, 4))
plt.plot(t, signal)
plt.title('Sample Signal (Sine Wave)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

# Compute FFT of the Signal
Compute the FFT of the sample signal using the defined FFT function.

```python
# Compute FFT of the Signal

def fft(signal):
    fft_signal = np.fft.fft(signal)
    return fft_signal

# Compute the FFT of the sample signal
fft_signal = fft(signal)

# Plot the FFT of the signal
plt.figure(figsize=(10, 4))
plt.plot(np.fft.fftfreq(len(fft_signal), 1/sampling_rate), np.abs(fft_signal))
plt.title('FFT of the Sample Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()

```

Mathematical Formulation of FFT

The FFT of the signal is computed using the following formula:
$$
X[k] = \sum_{n=0}^{N-1} x[n] \cdot e^{-j \frac{2\pi}{N} kn}, \quad k = 0, 1, \ldots, N-1
$$

# Visualize the FFT Result
Visualize the FFT result using Matplotlib.

```python
# Visualize the FFT Result

# Plot the FFT result (magnitude and phase)
plt.figure(figsize=(12, 6))

# Magnitude plot
plt.subplot(2, 1, 1)
plt.plot(np.fft.fftfreq(len(fft_signal), 1/sampling_rate), np.abs(fft_signal))
plt.title('FFT of the Sample Signal - Magnitude')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)

# Phase plot
plt.subplot(2, 1, 2)
plt.plot(np.fft.fftfreq(len(fft_signal), 1/sampling_rate), np.angle(fft_signal))
plt.title('FFT of the Sample Signal - Phase')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Phase [radians]')
plt.grid(True)

plt.tight_layout()
plt.show()

```

Mathematical Formulation of FFT Result

The magnitude of the FFT result is given by:
$$
|X[k]| = \sqrt{\text{Re}(X[k])^2 + \text{Im}(X[k])^2}
$$

The phase of the FFT result is given by:
$$
\text{Phase}(X[k]) = \tan^{-1}\left(\frac{\text{Im}(X[k])}{\text{Re}(X[k])}\right)
$$

# Compute Inverse FFT of the Signal
Compute the inverse FFT of the signal using the defined IFFT function.

```python
# Compute Inverse FFT of the Signal
def ifft(signal):
    ifft_signal = np.fft.ifft(signal)
    return ifft_signal


# Compute the inverse FFT of the signal
ifft_signal = ifft(fft_signal)

# Plot the inverse FFT of the signal
plt.figure(figsize=(10, 4))
plt.plot(t, ifft_signal.real)
plt.title('Inverse FFT of the Sample Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

```

Mathematical Formulation of IFFT

The IFFT of the signal is computed using the following formula:
$$
x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] \cdot e^{j \frac{2\pi}{N} kn}, \quad n = 0, 1, \ldots, N-1
$$

Visualize the inverse FFT result using Matplotlib.

```python
# Visualize the Inverse FFT Result

# Plot the inverse FFT result
plt.figure(figsize=(10, 4))
plt.plot(t, ifft_signal.real)
plt.title('Inverse FFT of the Sample Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

```
