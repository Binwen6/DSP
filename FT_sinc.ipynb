# Import Required Libraries
Import the necessary libraries, including NumPy and Matplotlib.

```python
# Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

# Define Functions
Define the sinc function and its Fourier Transform.

```python
# Define the sinc function
def sinc(t):
    return np.sinc(t / np.pi)

# Define the Fourier Transform of the sinc function
def fourier_transform_sinc(omega):
    return np.pi * np.sign(omega)

# Define the Inverse Fourier Transform of the sinc function
def inverse_fourier_transform_sinc(omega):
    return np.sinc(omega / np.pi)
```

# Compute Fourier Transform and Inverse Fourier Transform
Compute the Fourier Transform of the sinc function and then compute the inverse Fourier Transform.

```python
# Compute Fourier Transform and Inverse Fourier Transform

# Generate t and omega values
t_vals = np.linspace(-5, 5, 400)
omega_vals = np.linspace(-10, 10, 400)

# Compute the function values
f_t_vals = sinc(t_vals)
F_omega_vals = fourier_transform_sinc(omega_vals)
f_t_ift_vals = inverse_fourier_transform_sinc(omega_vals)

# Plot the sinc function, its Fourier Transform, and its Inverse Fourier Transform
plt.figure(figsize=(18, 6))

# Plot the original sinc function
plt.subplot(1, 3, 1)
plt.plot(t_vals, f_t_vals, label='$f(t) = \\mathrm{sinc}(t)$', color='blue')
plt.xlabel('$t$')
plt.ylabel('$f(t)$')
plt.title('Sinc Function')
plt.legend()
plt.grid(True)

# Plot the Fourier Transform of the sinc function
plt.subplot(1, 3, 2)
plt.plot(omega_vals, F_omega_vals, label='$F(\\omega) = \\pi \\cdot \\mathrm{sign}(\\omega)$', color='red')
plt.xlabel('$\\omega$')
plt.ylabel('$F(\\omega)$')
plt.title('Fourier Transform of Sinc Function')
plt.legend()
plt.grid(True)

# Plot the Inverse Fourier Transform of the sinc function
plt.subplot(1, 3, 3)
plt.plot(omega_vals, f_t_ift_vals, label='$f(t) = \\mathrm{sinc}(\\omega)$', color='green')
plt.xlabel('$\\omega$')
plt.ylabel('$f(t)$')
plt.title('Inverse Fourier Transform of Sinc Function')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
