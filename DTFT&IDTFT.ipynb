# Import Required Libraries
Import the necessary libraries, including NumPy and Matplotlib.

```python
# Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt
```

# Define DTFT Function
Define a function to compute the Discrete-Time Fourier Transform (DTFT) of a given signal. Use LaTeX to display the mathematical formula: $$X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j\omega n}$$

# Define IDTFT Function
Define a function to compute the Inverse Discrete-Time Fourier Transform (IDTFT) of a given DTFT. Use LaTeX to display the mathematical formula: $$x[n] = \frac{1}{2\pi} \int_{-\pi}^{\pi} X(e^{j\omega}) e^{j\omega n} d\omega$$

# Example Signal
Create an example discrete-time signal for demonstration purposes.

```python
# Create an example discrete-time signal for demonstration purposes
n = np.arange(-10, 11)
x = np.sin(0.2 * np.pi * n) + 0.5 * np.sin(0.6 * np.pi * n)

# Plot the example signal
plt.stem(n, x)
plt.title('Example Discrete-Time Signal')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid(True)
plt.show()
```

# Compute DTFT
Compute the DTFT of the example signal using the defined DTFT function.

```python
# Define the DTFT function
def dtft(x, n, omega):
    X = np.zeros(omega.shape, dtype=complex)
    for i, w in enumerate(omega):
        X[i] = np.sum(x * np.exp(-1j * w * n))
    return X

# Define the frequency range for DTFT computation
omega = np.linspace(-np.pi, np.pi, 1000)

# Compute the DTFT of the example signal
X = dtft(x, n, omega)
```

# Visualize DTFT
Visualize the magnitude and phase of the DTFT using Matplotlib.

```python
# Visualize DTFT

# Plot the magnitude of the DTFT
plt.figure()
plt.plot(omega, np.abs(X))
plt.title('Magnitude of DTFT')
plt.xlabel('Frequency (rad/sample)')
plt.ylabel(r'$|X(e^{j\omega})|$')
plt.grid(True)
plt.show()

# Plot the phase of the DTFT
plt.figure()
plt.plot(omega, np.angle(X))
plt.title('Phase of DTFT')
plt.xlabel('Frequency (rad/sample)')
plt.ylabel(r'$X(e^{j\omega})$')
plt.grid(True)
plt.show()
```

# Compute IDTFT
Compute the IDTFT of the DTFT using the defined IDTFT function.

```python
# Compute IDTFT

# Define the IDTFT function
def idtft(X, omega, n):
    x_reconstructed = np.zeros(n.shape, dtype=complex)
    for i, ni in enumerate(n):
        x_reconstructed[i] = (1 / (2 * np.pi)) * np.sum(X * np.exp(1j * omega * ni) * (omega[1] - omega[0]))
    return x_reconstructed

# Compute the IDTFT of the DTFT
x_reconstructed = idtft(X, omega, n)
```

# Visualize IDTFT
Visualize the reconstructed signal from the IDTFT and compare it with the original signal.

```python
# Visualize IDTFT

# Plot the reconstructed signal
plt.figure()
plt.stem(n, np.real(x_reconstructed))
plt.title('Reconstructed Signal from IDTFT')
plt.xlabel('n')
plt.ylabel(r'$Re{x[n]}$')
plt.grid(True)
plt.show()

# Plot the difference between the original and reconstructed signals
plt.figure()
plt.stem(n, np.real(x))
plt.title('Original Signals')
plt.xlabel('n')
plt.ylabel('Difference')
plt.grid(True)
plt.show()
```
