# Fourier Transformation and Its Inverse

In this notebook, we will derive the formula of Fourier Transformation for a non-periodic continuous-time signal and its inverse.


## Fourier Transformation

The Fourier Transform of a continuous-time signal $f(t)$ is defined as:

$$F(\omega) = \int_{-\infty}^{\infty} f(t) e^{-j\omega t} \, dt$$

where:
- $F(\omega)$ is the Fourier Transform of $f(t)$
- $\omega$ is the angular frequency
- $j$ is the imaginary unit
- $t$ is the time variable


## Inverse Fourier Transformation

The inverse Fourier Transform is used to convert the frequency domain representation back to the time domain. It is defined as:

$$f(t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} F(\omega) e^{j\omega t} \, d\omega$$

where:
- $f(t)$ is the original continuous-time signal
- $F(\omega)$ is the Fourier Transform of $f(t)$
- $\omega$ is the angular frequency
- $j$ is the imaginary unit
- $t$ is the time variable


## Example: Fourier Transform of a Gaussian Function

Let's consider an example where we compute the Fourier Transform of a Gaussian function $f(t) = e^{-t^2}$.


```python
# Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Define the symbolic variables
t, omega = sp.symbols('t omega')

# Define the Gaussian function
f_t = sp.exp(-t**2)

# Compute the Fourier Transform
F_omega = sp.integrate(f_t * sp.exp(-sp.I * omega * t), (t, -sp.oo, sp.oo))
F_omega.simplify()
```

```python
f_t_2 = 1 / (2 * np.pi) * sp.integrate(F_omega * sp.exp(sp.I * omega * t), (omega, -sp.oo, sp.oo))
f_t_2.simplify()
```

## Visualization

Let's visualize the original Gaussian function and its Fourier Transform.


```python
# Define the Gaussian function
def gaussian(t):
    return np.exp(-t**2)

# Define the Fourier Transform of the Gaussian function
def fourier_transform_gaussian(omega):
    return np.sqrt(np.pi) * np.exp(-omega**2 / 4)

# Generate t and omega values
t_vals = np.linspace(-5, 5, 400)
omega_vals = np.linspace(-10, 10, 400)

# Compute the function values
f_t_vals = gaussian(t_vals)
F_omega_vals = fourier_transform_gaussian(omega_vals)

# Plot the original Gaussian function
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(t_vals, f_t_vals, label='$f(t) = e^{-t^2}$')
plt.xlabel('$t$')
plt.ylabel('$f(t)$')
plt.title('Gaussian Function')
plt.legend()
plt.grid(True)

# Plot the Fourier Transform of the Gaussian function
plt.subplot(1, 2, 2)
plt.plot(omega_vals, F_omega_vals, label='$F(\omega) = \sqrt{\pi} e^{-\omega^2 / 4}$')
plt.xlabel('$\omega$')
plt.ylabel('$F(\omega)$')
plt.title('Fourier Transform of Gaussian Function')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
