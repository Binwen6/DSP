# Define DFT and IDFT Functions
Implement the DFT and IDFT functions using NumPy.

```python
# Define the DFT function
def DFT(x):
    """
    Compute the Discrete Fourier Transform of the 1D array x
    """
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    return np.dot(M, x)

# Define the IDFT function
def IDFT(X):
    """
    Compute the Inverse Discrete Fourier Transform of the 1D array X
    """
    X = np.asarray(X, dtype=complex)
    N = X.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(2j * np.pi * k * n / N)
    return np.dot(M, X) / N
```

# Mathematical Formulas for DFT and IDFT

## Discrete Fourier Transform (DFT)

The Discrete Fourier Transform (DFT) of a sequence $ x[n] $ of length $ N $ is given by:


$$X[k] = \sum_{n=0}^{N-1} x[n] \cdot e^{-j \frac{2\pi}{N} kn}$$

where $$ k = 0, 1, 2, \ldots, N-1 .$$

## Inverse Discrete Fourier Transform (IDFT)

The Inverse Discrete Fourier Transform (IDFT) of a sequence \( X[k] \) of length \( N \) is given by:

$$
x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] \cdot e^{j \frac{2\pi}{N} kn}
$$

where $$ n = 0, 1, 2, \ldots, N-1 $$

# Generate Sample Signal
Generate a sample signal (e.g., a sine wave) to demonstrate the DFT and IDFT.

```python
# Generate a sample signal: a sine wave
N = 128  # Number of sample points
T = 1.0 / 50.0  # Sample spacing
x = np.linspace(0.0, N*T, N, endpoint=False)
y = np.sin(2.0 * np.pi * x)

# Plot the sample signal
plt.figure(figsize=(10, 4))
plt.plot(x, y, label='Sample Signal (Sine Wave)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sample Signal')
plt.legend()
plt.grid()
plt.show()
```

# Compute and Visualize Original Signal and its DFT
Plot the original signal and its DFT using Matplotlib.

```python
# Visualize Original Signal and its DFT

# Plot the original signal
plt.figure(figsize=(10, 4))
plt.plot(x, y, label='Original Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Original Signal')
plt.legend()
plt.grid()
plt.show()

# Compute the DFT of the original signal
Y = DFT(y)

# Plot the magnitude of the DFT
plt.figure(figsize=(10, 4))
plt.plot(np.abs(Y), label='Magnitude of DFT')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('Magnitude of DFT')
plt.legend()
plt.grid()
plt.show()

# Plot the phase of the DFT
plt.figure(figsize=(10, 4))
plt.plot(np.angle(Y), label='Phase of DFT')
plt.xlabel('Frequency')
plt.ylabel('Phase (radians)')
plt.title('Phase of DFT')
plt.legend()
plt.grid()
plt.show()
```

# Compute IDFT of the Signal
Use the IDFT function to compute the inverse DFT of the transformed signal.

```python
# Compute IDFT of the Signal

# Compute the IDFT of the transformed signal
y_idft = IDFT(Y)

# Plot the IDFT of the transformed signal
plt.figure(figsize=(10, 4))
plt.plot(x, y_idft.real, label='IDFT of Transformed Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('IDFT of Transformed Signal')
plt.legend()
plt.grid()
plt.show()
```

# Visualize Reconstructed Signal
Plot the reconstructed signal and compare it with the original signal using Matplotlib.

```python
# Visualize Reconstructed Signal

# Plot the original signal
plt.figure(figsize=(10, 4))
plt.plot(x, y, label='Original Signal')
plt.plot(x, y_idft.real, label='Reconstructed Signal', linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Original Signal vs Reconstructed Signal')
plt.legend()
plt.grid()
plt.show()
```
